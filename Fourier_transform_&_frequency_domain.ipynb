```python
def plot_magnitude_spectrum(signal, title, sr, f_ratio = 1 ):
    ft = np.fft.fft(signal)
    magnitude_spectrum = np.abs(ft)
    
    #plot
    plt.figure(figsize=(15,5))
    
    frequency = np.linspace(0, sr, len(magnitude_spectrum))
    num_frequency_bin = int(len(frequency)*f_ratio)
    
    plt.plot(frequency[:num_frequency_bin],magnitude_spectrum[:num_frequency_bin])
    plt.xlabel("Frequency")
    plt.title(title)
    
    plt.show()
    
```

```python
plot_magnitude_spectrum(violin,'violin', sr,1)
```

```python
plot_magnitude_spectrum(sax,'sax', sr,0.1)
```

```python
plot_magnitude_spectrum(piano,'piano', sr,0.1)
```

```python
plot_magnitude_spectrum(noise,'noise', sr,0.5)
```
